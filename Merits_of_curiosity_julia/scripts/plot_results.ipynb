```julia
using LinearAlgebra
using MultivariateStats
using UMAP
using PyPlot
using DelimitedFiles

function plot_PCA(perf_filenames, out_filename, env_names; proj_type="PCA", pca_umap_path="")
    
    data_perf = load(perf_filenames[1])
    scores = data_perf["scores"]
    measures = data_perf["measures"]
    intr_types = data_perf["intr_types"]

    n_meas, n_intr, n_steps, n_repet = size(scores)
    n_envs = length(env_names)
    
    #data = [load(filename)["scores"] for filename in perf_filenames]
    data = []
    for filename in perf_filenames
        scores = load(filename)["scores"]
        auc = dropdims(mean(scores, dims=3), dims=3)
        push!(data, auc)
    end

    # Reshape the data
    samples = []
    labels = []
    for (env_idx, env_data) in enumerate(data)
        for rep in 1:n_repet
            sample = vec(env_data[:, :, rep])
            push!(samples, sample)
            push!(labels, env_idx)
        end
    end

    samples_matrix = hcat(samples...)
    
    writedlm( pca_umap_path*"samples_matrix.csv",  samples_matrix', ',')
    writedlm( pca_umap_path*"labels.csv",  labels, ',')

    projected_samples = nothing
    if proj_type == "PCA"
        # Perform PCA
        pca_model = MultivariateStats.fit(PCA, samples_matrix; maxoutdim=2)
        projected_samples = MultivariateStats.transform(pca_model, samples_matrix)'
        println(size(samples_matrix))
        println(size(projected_samples))
        println(size(labels))
    elseif proj_type == "UMAP"
        umap_model = UMAP_(samples_matrix, 2, metric="manhattan")#, min_dist=0.1)#, metric="euclidean")
        projected_samples = UMAP.transform(umap_model, samples_matrix)
    elseif proj_type == "LDA"
        lda_model = MultivariateStats.fit(MulticlassLDA, samples_matrix, labels; outdim=2)
        projected_samples = MultivariateStats.predict(lda_model, samples_matrix)'
    end
    # Plot the results using Pyplot
    colors = ["red", "blue", "green", "y", "pink"]  # Define different colors for each environment
    figure(figsize=(9, 6))
    for env_idx in 1:n_envs
        env_samples = projected_samples[labels .== env_idx, :]
        scatter(env_samples[:, 1], env_samples[:, 2], color=colors[env_idx], label=env_names[env_idx] * " env", s=50, alpha=0.6)
    end

    PyPlot.xlabel("Principal Component 1")
    ylabel("Principal Component 2")
    title("PCA Projection", fontsize=20)#title("Muli-class LDA", fontsize=20)# of Performance vectors from Different Environments")
    legend(fontsize=14)
    savefig(out_filename)
    close()
end
```

```julia
plot_name = "PCA"
setup_names = [
    "default_2000",
    "sink50_2000",
    "source50_2000",
    "stoc1_2000",
    "sink50-source50-stoc1_2000"
]

env_names = ["Neutral", "Sink", "Source", "Stochastic", "Mixed"]

perf_prefix = "" # path to the perf folder
plot_path = "" # path to the plot folder
perf_filenames = [perf_prefix*s*".jld" for s in setup_names]
out_filename = plot_path* plot_name *".pdf"
pca_umap_path = "" # path to the pca_umap folder, with the csv files

plot_PCA(perf_filenames, out_filename, env_names, proj_type="PCA", pca_umap_path=pca_umap_path)
```
